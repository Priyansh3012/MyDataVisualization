```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Top 20 most sold Categories**

```python
top_n = 20
category_counts = df['Description'].value_counts().head(top_n)
plt.figure(figsize=(12, 8))  # Adjust the figure size as needed
category_counts.plot(kind='bar')
plt.title(f'Top {top_n} Categories')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better spacing

top_categories = category_counts
for i, count in enumerate(category_counts):
    plt.text(i, count + 20, str(count), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

# **Top 20 most sold Categories(in percentage)**

```python
category_counts = df['Description'].value_counts()

# Select the top 20 categories
top_categories = category_counts.head(20)

# Create a pie chart
plt.figure(figsize=(15, 8))
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%')
plt.title('Top 20 most sold Categories\n')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Multivariate Analysis
```

# Most Sold Product in each Month

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract year and month from the 'date' column
df['year_month'] = df['InvoiceDate'].dt.to_period('M')

# Group by year and month, then find the most sold product
grouped = df.groupby(['year_month', 'Description'])['Quantity'].sum().reset_index()
idx = grouped.groupby('year_month')['Quantity'].idxmax()
most_sold_products = grouped.loc[idx]


plt.figure(figsize=(12, 8))
for year_month, group in most_sold_products.groupby('year_month'):
    bar_position = str(year_month)
    bar_height = group['Quantity'].values[0]
    plt.bar(bar_position, bar_height, label=group['Description'].values[0])
    plt.text(bar_position, bar_height + 1, group['Description'].values[0], ha='center', rotation=90, fontsize=9,
             verticalalignment='center')

plt.xlabel('Year-Month')
plt.ylabel('Quantity')
plt.title('Most Sold Product in each Month')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

# Most Sold Product in each Month (count)

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extracting year and month from the 'date' column
df['year_month'] = df['InvoiceDate'].dt.to_period('M')

# Group by year and month, then find the most sold product
grouped = df.groupby(['year_month', 'Description'])['Quantity'].sum().reset_index()
idx = grouped.groupby('year_month')['Quantity'].idxmax()
most_sold_products = grouped.loc[idx]


# Plotting
plt.figure(figsize=(15, 11))
for year_month, group in most_sold_products.groupby('year_month'):
    plt.bar(str(year_month), group['Quantity'], label=group['Description'].values[0])
    bar_height = group['Quantity'].values[0]
    plt.text(str(year_month), bar_height + 1, f"{group['Description'].values[0]}\n", ha='center', rotation=76,fontsize=10)
    plt.text(str(year_month), bar_height - 2, str(bar_height), ha='center', rotation=76,color='red',fontsize=15)

plt.xlabel('\nYear-Month')
plt.ylabel('Quantity')
plt.title('Most Sold Product in each Month (count)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

# Total Sale in each Month of the respective Year

```python
# Convert 'date' column to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Calculate total sale for each row
df['total_sale'] = df['Quantity'] * df['UnitPrice']

# Extract month and year from the 'date' column
df['month_year'] = df['InvoiceDate'].dt.to_period('M')

# Group by 'month_year' and calculate sum of 'total_sale'
monthly_sales = df.groupby('month_year')['total_sale'].sum()

# Create the bar chart
plt.figure(figsize=(12, 8))
ax = monthly_sales.plot(kind='bar')
plt.xlabel('Year and Month')
plt.ylabel('Total Sale')
plt.title('Total Sale in each Month of the respective Year')
plt.xticks(rotation=45)

# Customize y-axis ticks using np.arange with interval of 1000000
y_ticks = np.arange(0, monthly_sales.max() + 1000000, 1000000)
plt.yticks(y_ticks)

for i, v in enumerate(monthly_sales):
    ax.text(i, v + 50000, f'{v:,}', ha='center', va='bottom', fontsize=12,rotation=45,color='red')

plt.tight_layout()
plt.show()
```

# Top CustomerID of the Month and Number of Product Units they Purchased

```python
# CustomerID of the Month (most orders placed)

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract year and month from the 'date' column
df['year_month'] = df['InvoiceDate'].dt.to_period('M')

# Group by year and month, then find the most sold product
grouped = df.groupby(['year_month', 'CustomerID'])['Quantity'].sum().reset_index()
idx = grouped.groupby('year_month')['Quantity'].idxmax()
most_sold_products = grouped.loc[idx]


# Plotting
plt.figure(figsize=(15, 11))
for year_month, group in most_sold_products.groupby('year_month'):
    plt.bar(str(year_month), group['Quantity'], label=group['CustomerID'].values[0])
    bar_height = group['Quantity'].values[0]
    plt.text(str(year_month), bar_height + 1, f"{group['CustomerID'].values[0]}\n", ha='center', rotation=0,fontsize=15)
    plt.text(str(year_month), bar_height - 2, str(bar_height), ha='center', rotation=0,color='red',fontsize=12)

plt.xlabel('\nYear-Month')
plt.ylabel('Quantity')
plt.title('Top CustomerID of the Month (in BLACK) and Number of Product Units they Purchased (in RED).')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
